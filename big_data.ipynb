```python
# clustering the data 
import cluster

# defining parameters
k = 5
max_iter = 10

# randomly select only 10% of the data
music_csv = music_csv.sample(frac = 0.00001)

# calling clustering function
clusters = cluster.clustering(k, music_csv, max_iter)
```
